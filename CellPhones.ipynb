```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
import scipy.sparse
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
# Compute cosine similarity between all phones
#similarity_matrix = cosine_similarity(feature_matrix, feature_matrix)
#print(similarity_matrix)
# Define model name to search for
```

```python

# Load the data (Use `items` DataFrame)
df = items.copy()

# Combine relevant text features into a single string for each phone
df["features"] = df["brand"] + " " + df["model"] + " " + df["operating system"] + " " + \
                 df["internal memory"].astype(str) + "GB storage " + df["RAM"].astype(str) + "GB RAM " + \
                 df["main camera"].astype(str) + "MP main camera " + df["selfie camera"].astype(str) + "MP selfie camera " + \
                 df["battery size"].astype(str) + "mAh battery " + df["screen size"].astype(str) + "-inch screen " + \
                 df["weight"].astype(str) + "g weight " + "$" + df["price"].astype(str) + " price"

# Convert text features into numerical vectors using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english')
text_features = vectorizer.fit_transform(df["features"])

# Normalize numerical features (price, RAM, battery size)
scaler = StandardScaler()
numerical_features = scaler.fit_transform(df[["RAM", "battery size", "weight", "price", "internal memory", "performance", "screen size"]])

# Combine text and numerical features
combined_features = scipy.sparse.hstack((text_features, numerical_features)) #scipy.sparse.hstack() horizontally stacks (joins) text features and numerical features into one large feature matrix.
# Convert to indexable sparse matrix
combined_features = combined_features.tocsr()  # Convert to CSR format csr_matrix (Compressed Sparse Row Format),Efficient for row-based indexing (which we need).
knn = NearestNeighbors(n_neighbors=6, metric="cosine")  # K=6 (1 for itself + 5 recommendations)
knn.fit(combined_features)
```

```python
def recommend_knn_cosine(model_name, df, knn_model, feature_matrix, n=5):
    print("***********************************************")
    print(green("Content-Based Recommendations using KNN & Cosine Similarity:", ['bold']))
    print("***********************************************")
    print()
    
    # Convert model name to lowercase for consistency
    model_name = model_name.lower()

    # Check if the model exists in the dataset
    model_index = df[df["model"] == model_name].index
    if len(model_index) == 0:
        return f"Model '{model_name}' not found in the dataset."
    
    model_index = model_index[0]  # Take the first match if multiple exist
    display(Image(url=items['url'][model_index], width=250, height=250))
    print(red("Selected Phone:", ['bold']), items['model'][model_index])
    print(magenta("Price: $ {}".format(items['price'][model_index]), ['bold']))
    print(magenta("Features:", ['bold']), df['features'][model_index])
    print("==========================================================")


   
    # Ensure feature input is 2D for KNN
    distances, indices = knn_model.kneighbors(feature_matrix[model_index].reshape(1, -1), n_neighbors=n+1)

    # Get recommended phones (excluding the input phone itself)
    recommended_indices = indices.flatten()[1:]  # Exclude first entry (input model)

    
    print(red("Recommended Top {} Phones:".format(n), ['bold']))
    print()
    
    for indx in recommended_indices:
        display(Image(url=items['url'][indx], width=250, height=250))
        print(items['url'][indx])
        print(indx)
        
        print(red("Brand & Model:", ['bold']), items['brand'][indx], items['model'][indx])
        print(magenta("Price: $ {}".format(items['price'][indx]), ['bold']))
        print(magenta("Features:", ['bold']), df['features'][indx])
        print("------------------------------------------------------")
    

recommend_knn_cosine("iphone 13", df, knn, combined_features, n=5)
```

**Modeling**

 We need user rating matrix for modeling

```python
# Using pandas pivot table for user-item matrix formation

rui = pd.pivot_table(collabdata, index='user_id', columns='cellphone_id', values='rating').fillna(0)
```

```python
rui
```

* **bold text** from the above rui , we need to predict missed(na) rating values.
* to work on this we need to know how users are giving ratings to items
* this information will be given by laten factors, which are calculated from rui matrix

                    rui = userlatent matrix * latent factors * item(transpose)latent matrix
                    
* this can be imlemented in python using a module **surprise**

**Define & fit**

```python
from surprise import Dataset, Reader, SVD
from surprise.model_selection import train_test_split
from surprise import accuracy
```

```python
# Data Reading to Surprise
reader = Reader(rating_scale=(0.5, 5.0))  # Ratings range from 0.5 to 5.0
data = Dataset.load_from_df(interactions[['user_id', 'cellphone_id', 'rating']], reader)
```

```python
# Train-Test Split
trainset, testset = train_test_split(data, test_size=0.2, random_state=42)
```

```python
# Apply SVD
svd = SVD()
svd.fit(trainset)
```

```python
# Evaluate the model
predictions = svd.test(testset)
rmse = accuracy.rmse(predictions)
```

**Item Collaborative Recommendations:**

```python
# Creating  lookup dictionary
item_lookup = items.set_index("cellphone_id")["model"].to_dict()
```

```python
# Compute item latent vectors
item_factors = {trainset.to_raw_iid(itemid): svd.qi[itemid] for itemid in trainset.all_items()}
```

```python
# Function to recommend similar items

def recommend_similar_items(item_title, n=5):
    
    item_id = items[items["model"].str.contains(item_title, case=False, na=False)]["cellphone_id"]
    if item_id.empty:
        return []

    item_id = item_id.iloc[0]
    if item_id not in item_factors:
        return []

    item_vector = item_factors[item_id]
    # Compute cosine similarity with all other items
    similarities = {other_id: np.dot(item_vector, other_vector)
                    for other_id, other_vector in item_factors.items()
                    if other_id != item_id}

    # Get top N most similar items
    top_item_ids = sorted(similarities, key=similarities.get, reverse=True)[:n]
    
    return [(item_lookup[itemid], round(similarities[itemid],2)) for itemid in top_item_ids if itemid in item_lookup]
```

```python
# Generate item recommendations for a specific user

def recommend_items(userid, n=5):
    all_item_ids = interactions['cellphone_id'].unique()
    rated_items = interactions[interactions['user_id'] == userid]['cellphone_id'].values
    notrated_items = [item for item in all_item_ids if item not in rated_items]
    predictions = [(item, svd.predict(userid, item).est) for item in notrated_items]
    recommendations = sorted(predictions, key=lambda x: x[1], reverse=True)[:n]
    
    return recommendations

# Example: Recommend 5 movies for user 1
recommended_items = recommend_items(userid=100,n=5)
print("Top 5 Recommended Items for User :10")
print()
print(recommended_items)
```
